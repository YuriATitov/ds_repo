1. Какую часть зарплат от общей суммы получают менеждеры, а какую директора?

```python
pd.read_sql(
    """
SELECT distinct role--, sum(salary) over() as 'sum_all'
--,  sum(salary) over(partition by role) as 'sum_role'
, (sum(salary) over(partition by role))*100.0 / (sum(salary) over()) as 'prc_role'
FROM salaries
    """,
    con,
)
```

2. На каком месте по величине зарплаты находятся сотрудники, у которых первая буква в имени "R"? 

```python
pd.read_sql(
    """
select * from (
SELECT  emp, 
        salary, 
        dense_rank() over (order by salary) as 'sal_place_dr' 
FROM salaries 
) s
where s.emp like 'R%'
    """,
    con,
)
```

3. Выведите процент менеджеров и директоров по каждому из департаментов.

```python
pd.read_sql(
    """
select distinct depname, role 
--, count(role) over (partition by depname) as 'cnt_dep'
--, count(role) over (partition by depname,role) as 'cnt_role'

, (count(role) over (partition by depname,role))*100.0/(count(role) over (partition by depname)) as 'prc_role'
FROM salaries1

    """,
    con,
)
```

4. Выведите среднюю зарплату по 1 департаменту, а также по 2 и 3 совместно.

```python
pd.read_sql(
    """
with dept1 as (
select avg(salary)
from salaries
where depname = 'dep1'
),
dept23 as (
select avg(salary)
from salaries
where depname != 'dep1'
)
    
select 'dep1', * from dept1
union all
select 'dep23', * from dept23
    """,
    con,
)
```

```python
pd.read_sql(
    """
with dept1 as (
select depname,avg(salary)
from salaries
where depname = 'dep1'
union all
select depname,avg(salary)
from salaries
where depname != 'dep1'
)   
select * from dept1

    """,
    con,
)
```

5. Какой процент от суммы зарплат по всех организации получают 10% сотрудников с самыми большими зарплатами? 20%?30%?

```python
pd.read_sql(
    """
 select salary*100.0/sum_sal as 'prc_10'
 from (
select emp, salary, sum(salary) over () as 'sum_sal'
, ntile(100/10) over (order by salary desc) as '10_prc'
from salaries s
) ss
where ss."10_prc" = 1
    """,
    con,
)
```

```python
pd.read_sql(
    """
 select salary*100.0/sum_sal as 'prc_10'
 from (
select emp, salary, sum(salary) over () as 'sum_sal'
, ntile(10) over (order by salary desc) as '10_prc'
from salaries s
) ss
where ss."10_prc" <=2
    """,
    con,
)
```

```python
pd.read_sql(
    """
with table1 as (
select emp, salary, sum(salary) over () as 'sum_sal'
, ntile(10) over (order by salary desc) as '10_prc'
from salaries
)
select sum(salary)*100.0/ sum_sal as 'prc_20'  from table1
where "10_prc" <=3

    """,
    con,
)
```

```python
pd.read_sql(
    """
with table1 as (
select emp, salary, sum(salary) over () as 'sum_sal'
, ntile(100) over (order by salary desc) as '1_prc'
from salaries
)
select sum(salary)  from table1
where "1_prc" <=33

    """,
    con,
)
```
